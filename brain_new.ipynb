```python

from ultralytics import YOLO

import torch
```

```python
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
DEVICE
```

```python
model = YOLO("yolo11m.pt")
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("davidbroberts/brain-tumor-object-detection-datasets")

print("Path to dataset files:", path)
```

```python
results = model.train(
    data='/home/adminadmin/brain_dir/archive_b/axial_t1wce_2_class/axial_t1wce_2_class.yaml',
    epochs=80,  
    imgsz=640,  
    batch=8,  
    device=DEVICE,
    project='/home/adminadmin/brain_dir/dir_brain_tessssst',  
    name='train_exp1_axial', 
    patience=10,  
    save=True,  
    plots=True  
)
```

```python
url = "https://ihospital.ru/image/b99d358d-e196-4120-b98f-349a163abf9f.png"

results = model.predict(url, conf=0.1)
```
